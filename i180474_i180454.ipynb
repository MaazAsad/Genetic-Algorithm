#Exam struct to make chromosome

```python
class Exam:
  def __init__(self, Teacher,Classroom,StartT,Day,Course):

    self.Invigilator = Teacher
    self.Classroom = Classroom
    self.StartTime = StartT
    self.Day = Day
    self.Course = Course
  
  def Chromosome(self):
    c = tuple((self.Invigilator[0],self.Course[0]))
    # Chromo = str(self.Invigilator[0]) + " " + str(self.Course[0])
    return c
  
      
```

#Population structure

```python
class Schedule: 
  def __init__(self,x=1):
    self.row = 20
    self.col = 10 # classrooms
    self.table = [self.col*[0] for i in range(self.row)]
    
    self.HardFitness = 0.0 
    self.SoftFitness = 0.0
    self.TotalFitness = 0.0 
    self.populate()
    if(x==1):
      self.Fittness()
  
  def printTable(self):
    for i in range(self.row):
      for j in range(self.col):
        print(self.table[i][j], end=' ')
      print()
    print('\n')
  
  def getExam(self):
    TeacherChoice = random.choice(list(Teachers.keys()))
    ClassRoomChoice = random.choice(list(ClassRooms.keys()))
    TimeSlot = random.choice(list(Time.keys()))
    DayChoice = random.choice(list(Days.keys()))
    CourseChoice = random.choice(list(Courses.keys()))

    t = (TeacherChoice,Teachers[TeacherChoice])
    classRoom = (ClassRoomChoice,ClassRooms[ClassRoomChoice])
    time = (TimeSlot,Time[TimeSlot])
    day = (DayChoice,Days[DayChoice])
    c = (CourseChoice,Courses[CourseChoice])
    tempExam = Exam(t,classRoom,time,day,c)
    return tempExam
  
  def getStudentList(self, CourseName)-> list():
    to_ret = []
    for i in Students_Registration:
      if(CourseName in Students_Registration[i]):
        to_ret.append(i)
    return to_ret
  
  def populate(self):
    loop = 0
    cour = list()
    while (loop < len(Courses)):
      tempExam = self.getExam()
      gene = tempExam.Chromosome()
      
      r = 0
      if tempExam.StartTime[0] == "0":
        r = int(tempExam.Day[0],2) * 2 -2
      elif tempExam.StartTime[0] == "1":
        r = int(tempExam.Day[0],2) * 2 -1
      c = int(tempExam.Classroom[0],2) - 1 
      if tempExam.Course[0] not in cour:
        if self.table[r][c] == 0:
          cour.append(tempExam.Course[0])
          self.table[r][c] = gene
          loop+=1
    # self.printTable()
      
  def Fittness(self):
    self.HardFitness =0.0
    self.HardFitness += self.allCoursesScheduled() 
    self.HardFitness += self.StudentClashes()
    self.HardFitness += self.WeekendExamsCheck()
    self.HardFitness += self.ExamTiming()
    self.HardFitness += self.InvigilatorClash()
    self.HardFitness += self.InvigilatorDuty()

    self.SoftFitness= 0.0
    self.SoftFitness += self.FridayBreak()
    self.SoftFitness += self.ConsecutiveExam()
    self.SoftFitness += self.CoursePriority()
    self.SoftFitness += self.TwoHourBreak()

    self.TotalFitness = 100 - (self.HardFitness + self.SoftFitness)

   
  #HARD CONSTRAINTS CALCULATIONS
  def allCoursesScheduled(self)-> float:
    x=0.0
    CourseList = list(Courses.values())
    CoursesAppended = []
    for i in range(self.row):
      for j in range(self.col):
        if(self.table[i][j] != 0 ):
          if(Courses[self.table[i][j][1]] in CourseList and Courses[self.table[i][j][1]] not in CoursesAppended):
            CoursesAppended.append(Courses[self.table[i][j][1]]) 
            x+=1.0
    x = len(Courses) - x
    return x/len(Courses) * 14

  def StudentClashes(self)->float:
    x=0.0
    for i in range(self.row):
      for j in range(self.col):
        if(self.table[i][j] !=0):
          CourseScheduled = Courses[self.table[i][j][1]]
          StudentList = self.getStudentList(CourseScheduled)

          for L in StudentList:
            count =0
            for k in range(self.col):
              if (self.table[i][k] !=0 and k!=j):
                CourseToCheck = Courses[self.table[i][k][1]]
                StudentsToCheck = self.getStudentList(CourseToCheck)
                
                if L in StudentsToCheck:
                  count +=1
            NCourses = len(Students_Registration[L])-1
            count /= NCourses
            
            x+=count
    
    return x/len(Students_Registration) * 14
  
  def WeekendExamsCheck(self)->float:
    x = 0.0
    Weekendlist = []
    for i in Days:
      if(Days[i] == "Sat" or Days[i] == "Sun"):
        Weekendlist.append(i)

    for i in Weekendlist:
      MorningRow = int(i,2)*2-2
      AfternoonRow = int(i,2)*2-1
      for j in range(self.col):
        if(self.table[MorningRow][j] != 0):
         x+=1.0
         break

      for j in range(self.col):
        if(self.table[AfternoonRow][j] !=0):
          x+=1.0
          break
           
    return x/self.row * 14
 
  def ExamTiming(self)->float:
    x=0
    CoursesScheduled = 0
    MorningTime = Time["0"]
    MorningEndTime = int(MorningTime) + 300

    EveningTime = Time["1"]
    EveningEndTime = int(EveningTime) + 300

    for i in range(self.row):
      for j in range(self.col):
        if(self.table[i][j] !=0):
          CoursesScheduled+=1
          if(i%2 ==0 ):
            TempMorning = Time["0"]
            if(TempMorning != MorningTime and int(TempMorning) + 300 != MorningEndTime):
              x+=1.0
          else:
            TempEvening = Time["1"]
            if(TempEvening != EveningTime and int(TempEvening) + 300 != EveningEndTime):
              x+=1.0

    return x/(CoursesScheduled)*14
  
  def InvigilatorClash(self)->float:
    x=0.0
    MegaTeacherList = []
    for i in range(self.row):
      TeacherList = []
      for j in range(self.col):
        if(self.table[i][j] !=0):
          Invigilator = Teachers[self.table[i][j][0]]
          if(Invigilator not in MegaTeacherList):
            MegaTeacherList.append(Invigilator)
          if(Invigilator not in TeacherList):
            TeacherList.append(Invigilator)
            Nexams = 0
            CountTeacherViolations = 0
            for k in range(self.col):
              if(self.table[i][k] !=0 and k !=j):
                Nexams +=1
                if(Invigilator == Teachers[self.table[i][k][0]]):
                  CountTeacherViolations+=1
            if(Nexams != 0 and CountTeacherViolations !=0):
              x+= CountTeacherViolations/Nexams

    return x/len(MegaTeacherList) * 14

  def InvigilatorDuty(self)->float:
    x=0.0
    MegaTeacherList = []
    for i in range(self.row):
      for j in range(self.col):
        if(self.table[i][j] !=0):
          Invigilator = Teachers[self.table[i][j][0]]
          if(Invigilator not in MegaTeacherList):
            MegaTeacherList.append(Invigilator)
          if(i%2 == 0):
            nextTimeSlot = i+1 #afternoon slot
            Nexams = 0
            CountTeacherViolations = 0
            for k in range(self.col):

              if(self.table[nextTimeSlot][k] !=0):
                Nexams +=1
                if(Invigilator==Teachers[self.table[nextTimeSlot][k][0]]):
                  CountTeacherViolations +=1
                  
            if(Nexams != 0):
              x+= CountTeacherViolations/Nexams
    return x/len(MegaTeacherList) * 14

  ############################################################3
  # SOFT CONSTRAINTS CALCULATIONS
  def FridayBreak(self)->float:
    return 0.0
  def ConsecutiveExam(self)->float:
    x=0.0
    for i in range(self.row):
      for j in range(self.col):
        StdList = []
        if(self.table[i][j] !=0 and i%2 ==0):
          c = Courses[self.table[i][j][1]]
          StdList = self.getStudentList(c)

          for L in StdList:
            count =0
            for k in range(self.col):
              if (self.table[i+1][k] !=0 and k!=j):
                CourseToCheck = Courses[self.table[i+1][k][1]]
                StudentsToCheck = self.getStudentList(CourseToCheck)
                
                if L in StudentsToCheck:
                  count +=1
            NCourses = len(Students_Registration[L])-1
            count /= NCourses
            x+=count

    return x/len(Students_Registration) * 4

  def CoursePriority(self)->float:
    CsCourses = []
    MgCourses = 0
    tempVar = 0
    V = 0.0
    for i in Courses:
      if(Courses[i][0:2:1] == "CS"):
        CsCourses.append(i)
    
    for studs in Students_Registration:
      Studs_Courses = Students_Registration[studs]
      count =0
      MgCourses = 0
      tempVar = 0
      for x in Studs_Courses:

        if(x[0:2:1] == "MG"):
          MG_Course_Key = ""
          MgCourses +=1
          for key_, Val in Courses.items(): 
            if x == Val:
                MG_Course_Key = key_
          x=0
          y=0
          breakLoops = False
          for i in range(self.row):
            for j in range(self.col):
              if(self.table[i][j] != 0):
                if(self.table[i][j][1] == MG_Course_Key):
                  x,y = i,j
                  breakLoops = True
                  break
            if(breakLoops):
              break
          for i in range(0,x):
            for j in range(self.col):
              if(self.table[i][j] !=0):
                if(self.table[i][j][1] in CsCourses):
                  count +=1
                  tempVar +=1
        if(MgCourses !=0 and tempVar !=0):
          count /= (MgCourses*tempVar)
          V+= count

    return V/len(Students_Registration) * 4

  def TwoHourBreak(self)->float:
    x=0.0
    MorningTime = Time["0"]
    MorningEndTime = int(MorningTime) + 300

    EveningTime = Time["1"]
    EveningEndTime = int(EveningTime) + 300
    
    for i in range(self.row):
      for j in range(self.col):
        if(i%2 == 0):
          TempMorning = Time["0"]

          if(int(TempMorning)+300 > MorningEndTime and int(TempMorning)+300 < int(EveningTime)):
            x+=1
        else:
          TempEvening = Time["1"]
          if(int(TempEvening)+300 < MorningEndTime and int(TempEvening)+300 > int(EveningTime)):
            x+=1

    return x/len(Teachers)*4

  #########################################################
  #Display Fitness results
  def PrintFitnesses(self):
    print("TotalFitness: ", self.TotalFitness)
    print("Hard Constraints: ",self.allCoursesScheduled(),' ',self.StudentClashes(),' ',self.WeekendExamsCheck(),' ',self.ExamTiming(),' ',self.InvigilatorClash(),' ',self.InvigilatorDuty() )
    print("Soft Constraints: ", self.FridayBreak(),' ',self.ConsecutiveExam(),' ',self.CoursePriority(),' ',self.TwoHourBreak())


  def ConstraintFulfilled(self):
    countHard = 0
    countSoft = 0

    print("TotalFitness: ", self.TotalFitness)

    print("\nHard Constraints Fulfulled: ")
    print("Hard Constraint Fitnessvalue: ", 84- self.HardFitness, "/84 \n")
    if(self.allCoursesScheduled() == 0):
      print("All Courses Scheduled Constraint Fullfilled")
      countHard+=1
    else:
      print("All Courses Scheduled Constraint not Fullfilled, Constraint Value: ",self.allCoursesScheduled())

    if(self.StudentClashes() == 0):
      print("Student Clashes Constraint Fullfilled")
      countHard+=1
    else:
      print("Student Clashes Constraint not Fullfilled, Constraint Value: ",self.StudentClashes())
    
    if(self.WeekendExamsCheck() == 0):
      print("Weekend Exams Constraint Fullfilled")
      countHard+=1
    else:
      print("Weekend Exams Constraint not Fullfilled, Constraint Value: ",self.WeekendExamsCheck())

    if(self.ExamTiming() == 0):
      print("Exam Timing Constraint Fullfilled")
      countHard+=1
    else:
      print("Exam Timing Constraint not Fullfilled, Constraint Value: ",self.ExamTiming())
    
    if(self.InvigilatorClash() == 0):
      print("Invigilator Clash Constraint Fullfilled")
      countHard+=1
    else:
      print("Invigilator Clash Constraint not Fullfilled, Constraint Value: ",self.InvigilatorClash())
    
    if(self.InvigilatorDuty() == 0):
      print("Invigilator Duty Constraint Fullfilled")
      countHard+=1
    else:
      print("Invigilator Duty Constraint not Fullfilled, Constraint Value: ",self.InvigilatorDuty())
    

    print("\nSoft Constraints Fulfulled: ")
    print("Soft Constraint Fitnessvalue: ", 16- self.SoftFitness, "/16 \n")
    if(self.FridayBreak() == 0):
      print("Friday Break Constraint Fullfilled")
      countSoft +=1
    else:
      print("Friday Break Constraint not Fullfilled, Constraint Value: ",self.FridayBreak())
    
    if(self.ConsecutiveExam() == 0):
      print("Consecutive Exam Constraint Fullfilled")
      countSoft +=1
    else:
      print("Consecutive Exam Constraint not Fullfilled, Constraint Value: ",self.ConsecutiveExam())
    
    if(self.CoursePriority() == 0):
      print("Course Priority Constraint Fullfilled")
      countSoft +=1
    else:
      print("Course Priority Constraint not Fullfilled, Constraint Value: ",self.CoursePriority())
    
    if(self.TwoHourBreak() == 0):
      print("TwoHour Break Constraint Fullfilled")
      countSoft +=1
    else:
      print("TwoHour Break Constraint not Fullfilled, Constraint Value: ",self.TwoHourBreak())
    
    
    print()
    print("Number of Hard Constraints Fulfilled = ",countHard,"/6")
    print("Number of Soft COnstraints Fulfilled = ",countSoft,"/4")
```

#Roulette Wheel Selection using Probability and Cummulative Probability

```python
def SumOfFitness(Schedules) -> float:
  Sum_Fitness = 0.0
  for i in Schedules:
    Sum_Fitness += Schedules[i].TotalFitness
  # Sum_Fitness = "{:.6f}".format(Sum_Fitness)
  return float(Sum_Fitness)

def Probability(Schedules)->list():
  sum = SumOfFitness(Schedules)
  to_ret = list()
  for i in Schedules:
    tempP = Schedules[i].TotalFitness/sum
    # tempP = "{:.7f}".format(tempP)
    # tempP = float(tempP)
    to_ret.append(tempP)
  
  return to_ret

def CummProb(Prob) ->list():
  to_ret = list()
  runningSum = 0.0
  for i in Prob:

    runningSum += i
    runningSum = "{:.7f}".format(runningSum)
    runningSum = float(runningSum)
    to_ret.append(runningSum)
  return to_ret


def RouletteSelection(Pop,CumProb):
  tempSch = None

  while(tempSch == None):
    r1 = random.uniform(0,1)
    r1 = "{:.7f}".format(r1)
    r1 = float(r1)
    for i in range(len(CumProb)-1):
      if(i == 0):
        if(r1 < CumProb[0]):
          tempSch = Pop[i]
          break
      elif(r1 >CumProb[i] and r1< CumProb[i+1]):
        tempSch = Pop[i+1]
        break
    

  return tempSch
```

#Crossover and Mutation Function

**Different Crossover Functions**

* Single Point
* 2 point
* Uniform

```python
#Uniform crossover
def Crossover(sch1,sch2):
  nChanges = random.randrange(3,7)
  colSwapped = []
  for i in range(nChanges):
    ColToSwap = random.randrange(0,10)
    while (ColToSwap in colSwapped):
      ColToSwap = random.randrange(0,10) # get a row that hasnot been swapped already
    colSwapped.append(ColToSwap)
    for j in range(sch1.row):
      sch1.table[j][ColToSwap] , sch2.table[j][ColToSwap] = sch2.table[j][ColToSwap] , sch1.table[j][ColToSwap]
  
  return sch1,sch2

#Two Point Crossover
def Crossover1(sch1,sch2):
  ColStart = random.randrange(2,5)
  ColEnd = random.randrange(6,9)
  for j in range(sch1.row):
    for k in range(ColStart,ColEnd+1,1):
      sch1.table[j][k] , sch2.table[j][k] = sch2.table[j][k] , sch1.table[j][k]
  
  return sch1,sch2

#Single Point Crossover
def Crossover2(sch1,sch2):
  ColStart = random.randrange(4,7)
  for j in range(sch1.row):
    for k in range(ColStart,sch1.col,1):
      sch1.table[j][k] , sch2.table[j][k] = sch2.table[j][k] , sch1.table[j][k]
  
  return sch1,sch2
```

**Mutation Function**
* Random resetting

```python
def Mutation(sch1):
  
  newSch = Schedule(0)

  daytoSwap = random.choice(list(Days.keys()))

  row = int(daytoSwap,2)
  morningSLot = row*2-2
  eveningSlot = row*2-1

  for i in range(sch1.col):
    sch1.table[morningSLot][i] = newSch.table[morningSLot][i]
    sch1.table[eveningSlot][i] = newSch.table[eveningSlot][i]

  return sch1

```

#Genetic Algorithm

**Using 200 generations**

```python
CrossoverProb = 0.3
MutationProb = 0.3
Threshold = 80

NumberGen=200
nCarry = 10

for i in range(NumberGen):
  NewPop = dict()

  ProbList = Probability(Population)
  CumProbList = CummProb(ProbList)
  
  ##insert top 10 Fittest Chromosomes from previous Generation
  my_objs = sorted(Population.values(), key=operator.attrgetter('HardFitness'), reverse=False)[:nCarry]
  NChild = 0
  for x in range(nCarry):
    NewPop[NChild] = my_objs[x]
    NChild +=1

  s1 = RouletteSelection(Population,CumProbList)
  s2 = RouletteSelection(Population,CumProbList)
  

  while(len(NewPop) < len(Population)):
    s3 = copy.deepcopy(s1)
    s4 = copy.deepcopy(s2)

    rCross = random.randrange(0,1)
    rCross = "{:.1f}".format(rCross)
    rCross = float(rCross)

    rMutate = random.randrange(0,1)
    rMutate = "{:.1f}".format(rMutate)
    rMutate = float(rMutate)

   
    if(rCross <= CrossoverProb):
      s3,s4 = Crossover1(s3,s4)
      
   
    
    if(rMutate <= MutationProb):
      s3,s4 = Mutation(s3),Mutation(s4)
      
    startT = time.time()
    s3.Fittness()
    s4.Fittness()

    # if((84-s3.HardFitness) >= Threshold):
    #   NewPop[NChild] = copy.deepcopy(s3)
    #   NChild +=1
    
    # if((84-s4.HardFitness) >= Threshold):
    #   NewPop[NChild] = copy.deepcopy(s4)
    #   NChild +=1
    
    
    if(s3.TotalFitness > s4.TotalFitness):
      NewPop[NChild] = s3
      NChild +=1
    else:
      NewPop[NChild] = s4
      NChild+=1
   

    
    if(84-s1.HardFitness < 84-s2.HardFitness):
      s1 = RouletteSelection(Population,CumProbList)
      
    else:
      s2 = RouletteSelection(Population,CumProbList)
    
   
  
  Population = copy.deepcopy(NewPop)
  Threshold +=0.2
  
  print("Generation #:",i+1)
  # #Print Every 20th Generation
  # if((i+1)%20 == 0):
  #   for j in Population:
  #     print(84- Population[j].HardFitness)

```
